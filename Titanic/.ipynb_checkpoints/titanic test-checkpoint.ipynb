```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
## Display all the columns of the dataframe

pd.pandas.set_option('display.max_columns',None)
pd.pandas.set_option('display.max_rows',None)
```

```python
test = pd.read_csv("test.csv")
```

```python
test.head()
```

```python
test.drop(["PassengerId","Name"],axis =1, inplace = True)
```

```python
test.head(20)
```

```python
test.isnull().sum()
```

```python
random = test["Age"].dropna().sample(test["Age"].isnull().sum(),random_state =0)
```

```python
random.index = test[test["Age"].isnull()].index
```

```python
random.index
```

```python
test["Age_random"] = test["Age"]
```

```python
test.loc[test["Age"].isnull(),"Age_random"]=random
```

```python
test.head()
```

```python
test["Fare"] = np.log1p(test["Fare"])
```

```python
test.drop(["Ticket","Embarked"],axis =1, inplace = True)
```

```python
test["Cabin"] = test["Cabin"].astype(str).str[0]
```

```python
test["Cabin"].value_counts()
```

```python
test.isnull().sum()
```

```python
test.drop("Age",inplace = True,axis =1)
```

```python
test.head()
```

```python
dummy_test = pd.get_dummies(test,columns=["Sex","Cabin"])
```

```python
dummy_test["Cabin_T"] = np.zeros(shape=(418,1))
```

```python
dummy_test["Fare"] = dummy_test["Fare"].fillna(dummy_test["Fare"].median())
```

```python
dummy_test.isnull().sum()
```

```python
dummy_test =dummy_test[['Pclass', 'SibSp', 'Parch', 'Fare', 'Age_random',
       'Sex_female', 'Sex_male', 'Cabin_A', 'Cabin_B', 'Cabin_C', 'Cabin_D',
       'Cabin_E', 'Cabin_F', 'Cabin_G', 'Cabin_T', 'Cabin_n']]
```

```python
dummy_test.drop("Cabin_n",axis=1, inplace = True)
```

```python
dummy_test.head()
```

```python
dummy_test.shape
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler=MinMaxScaler()
scaler.fit(dummy_test)
```

```python
final_test = scaler.transform(dummy_test)
```

```python
final_test = pd.DataFrame(final_test,columns=dummy_test.columns)
```

```python
cont_feat = [feat for feat in final_test.columns if final_test[feat].dtypes !="O"]
```

```python
for i in cont_feat:
    if i!="Survived":
        final_test[i]=np.log1p(final_test[i])
    else:
        pass
```
