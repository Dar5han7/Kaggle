```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
## Display all the columns of the dataframe

pd.pandas.set_option('display.max_columns',None)
pd.pandas.set_option('display.max_rows',None)
```

```python
house_t = pd.read_csv("test.csv")
```

```python
house_t.head()
```

```python
house_t.info()
```

```python
house_t.isnull().sum()
```

```python
cont_feat = [feat for feat in house_t.columns if house_t[feat].isnull().sum()>0 and house_t[feat].dtypes!="O" and "Yr" not in feat]
cont_feat
```

```python
for items in cont_feat:
    if items !='LotFrontage':
        house_t[items] = house_t[items].fillna(house_t[items].median())
```

```python
sample = house_t["LotFrontage"].dropna().sample(house_t["LotFrontage"].isnull().sum(),random_state=1)
```

```python
house_t["LotFrontage_random"] = house_t["LotFrontage"]
```

```python
sample.index = house_t[house_t["LotFrontage"].isnull()].index
```

```python
house_t.loc[house_t["LotFrontage"].isnull(),"LotFrontage_random"] = sample
```

```python
house_t.head(10)
```

```python
fig = plt.figure(figsize=(20,10))
ax = fig.add_subplot(121)
house_t['LotFrontage'].plot(kind='kde', ax=ax)
house_t.LotFrontage_random.plot(kind='kde', ax=ax, color='green')
lines, labels = ax.get_legend_handles_labels()
ax.legend(lines, labels, loc='best')
```
